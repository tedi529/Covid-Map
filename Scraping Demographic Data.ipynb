```python
import pandas as pd
```

```python
zip_url = "http://zipatlas.com/us/zip-code-comparison/population-density.htm"
```

```python
# Scrape website to get dataframe
zips = pd.read_html(zip_url)
zips_df = zips[10]
zips_df
```

```python
# Create URLs list
urls =[]

for counter in range (2, 319) :
    url = "http://zipatlas.com/us/zip-code-comparison/population-density." + str(counter) + ".htm" 
    urls.append(url)
```

```python
# Loop through URLs to extract information
for url in urls:
    
    zip_codes = pd.read_html(url)
    zip_codes_df = zip_codes[10]
    zips_df = zips_df.append(zip_codes_df)

print(zips_df)
```

```python
# Save information as xls file
zips_df.to_excel("zips.xlsx")
```
